```python
response = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": ""},
         {"role": "assistant", "content": """
        JavaScriptTypeScript
        """},
        {"role": "user", "content": """
        JavaScriptTypeScript
        
        const fs = require('fs');
const path = require('path');
var sdk = require("microsoft-cognitiveservices-speech-sdk");
var readline = require("readline");

require('dotenv').config()

const subscriptionKey = process.env.TEXT2SPEECH_KEY
const subscriptionRegion = TEXT2SPEECH_REGION;

const filename = process.argv[2]

//.wav
async function speakTextAsync(filename) {

  //
  
  var extname = path.extname(filename);
  var basename = path.basename(filename, extname);

  // 
  var audioFile = basename + ".wav";
  
  audioConfig = sdk.AudioConfig.fromAudioFileOutput(audioFile);

  // This example requires environment variables named "SPEECH_KEY" and "SPEECH_REGION"
  const speechConfig = sdk.SpeechConfig.fromSubscription(subscriptionKey, subscriptionRegion);

  // The language of the voice that speaks.
  speechConfig.speechSynthesisVoiceName = "zh-CN-YunyangNeural";

  // Create the speech synthesizer.
  var synthesizer = new sdk.SpeechSynthesizer(speechConfig, audioConfig);

  var rl = readline.createInterface({
    input: process.stdin,
    output: process.stdout
  });


  let content = fs.readFileSync(filename, 'utf-8')
  const audioConfig = sdk.AudioConfig.fromAudioFileOutput(filename + ".wav");
  synthesizer.speakTextAsync(content, function (result) {
    if (result.reason === sdk.ResultReason.SynthesizingAudioCompleted) {
      console.log("completed.")

    } else {
      console.error("error" + result.errorDetails)
    }
    synthesizer.close();
    synthesizer = null;
  })
}        
        """}
    ]
)
```

```python
print(response.choices[0].message.content)
```
